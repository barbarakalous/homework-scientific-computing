```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[0, 0],
                   [1, 1],
                   [-1, -1],
                   [0, 1],
                   [0, 0.7],
                  ]).T
print(points)
```

```python
angle = np.radians(90)
theta_cos = np.cos(angle)
theta_sin = np.sin(angle) 

rotation_matrix = np.array([[theta_cos, -theta_sin],
                           [theta_sin, theta_cos]])

rotated_points = rotation_matrix @ points 
```

The result should like like this:

```python
plt.plot(*points, 'o', label='original points')
plt.plot(*rotated_points, 'o', label= 'rotated points')
plt.legend()
```

```python
plt.plot(*points, 'o', label='original points')
plt.plot(*rotated_points, 'o', label='rotated points')
plt.gca().set_aspect('equal')
```

### Exercise 3 - Flatten the curve

Copy the function `new_infections(t, k)` from last week's homework (exercise 3) and re-do the exercise using Numpy arrays instead of Python lists.

What needs to be changed in the function `new_infections(t, k)` to make this work?

```python
time = np.arange(251) # instead of: time = [x for x in range(0, 250+1)]
P = 1000000
k = 3/(10*P)


def new_infections(t, k ):
    i0 = 1  
    result = (((np.exp(-k*P*t))*k*(P**2)*(-1+(P/i0)))/(1+(np.exp(-k*P*t))*(-1+(P/i0)))**2) #using np.exp instead of math.exp and we don't need to create a new arraay  
    return result

flatten = new_infections(time,k,)
plt.plot(flatten, label = 'new infections')
plt.axhline(y = 20000, color = 'r')

plt.xlabel('time')
plt.ylabel('people')

plt.title('flatten the curve')
         
plt.legend()
plt.show() 
```
