### Exercise 2 - rotate and plot points in 2D

Plot the 5 points in 2D defined in the array `points`, then rotate the points by 90 degrees by performing a matrix multiplication with a [rotation matrix](https://en.wikipedia.org/wiki/Rotation_matrix) by using `rotation_matrix @ points` and plot the result in the same plot. The rotation angle needs to be converted to radians before it is passed to `np.cos()` and `np.sin()`, use `np.radians(90)` to do so.

```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[0, 0],
                   [1, 1],
                   [-1, -1],
                   [0, 1],
                   [0, 0.7],
                  ]).T
print(points)
```

```python
angle = np.radians(90)
sinus = np.sin(angle)
cosinus = np.cos(angle)
 
rotation_matrix = np.array([[cosinus, -sinus],
                           [sinus, cosinus]])

rotated_points = rotation_matrix @ points
```

The result should like like this:

```python
plt.plot(*points, "o", label="original points", color = "red")
plt.plot(*rotated_points, "o", label= "rotated points", color = "green")
plt.gca().set_aspect("equal")


plt.legend()
```

### Exercise 3 - Flatten the curve

What needs to be changed in the function `new_infections(t, k)` to make this work?

```python
t = np.arange(200)
P = 1000000
k = 3 / (P * 10)

def new_infections(t, k):
    infection_number = ((e**(-k*P*t))*k*P**2*(-1+(P/1))) / ((1+(e**(-k*P*t))*(-1+(P/1)))**2)
    return infection_number

plt.plot(t,new_infections(t,k),label = "infected", color = "green")
plt.axhline(y=30000, xmin=0, xmax=0.9, linewidth=2, color = "r", label="max healthcare system capacity")

plt.xlabel("days")
plt.ylabel("number of infections")
plt.title("Infection without any healtcare measures")

plt.legend()
plt.show()

k1 = np.array([3/(P*10),2/(P*10),1/(P*10)])

for l in k1:
    plt.plot(t,new_infections(t,l), label="infected")
 
plt.axhline(y=30000, xmin=0, xmax=0.9, linewidth=2, color = "r", label="max healthcare system capacity")

plt.xlabel("days")
plt.ylabel("number of infections")
plt.title("Infection with healtcare measures")

plt.legend()
plt.show()
```
